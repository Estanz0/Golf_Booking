# GolfBookingApp


```python
from selenium import webdriver
import yaml
from datetime import datetime, timedelta
from collections import namedtuple
import time

other_players = []
number_of_holes = 18

driver = webdriver.Chrome()

# ----------------- Login -----------------#

conf = yaml.load(open('venv/loginDetails.yaml'), Loader=yaml.FullLoader)
username = conf['golf_user']['id']
password = conf['golf_user']['password']

url = "https://www.golf.co.nz/my-golf-login"
usernameId = "ctl39_tbMembershipNumber"
passwordId = "ctl39_btnLogin"
submit_buttonId = "ctl39_btnLogin"

driver.get(url)
driver.find_element_by_id(usernameId).send_keys(username)
driver.find_element_by_id(passwordId).send_keys(password)
driver.find_element_by_id(submit_buttonId).click()

# ----------------- Navigate to Bookings -----------------#
book_a_tee_xpath = "//a[@class='site-header__action action--book-tee']"

while driver.current_url != 'https://www.golf.co.nz/Teebooking/SearchClubDay.aspx':
    driver.find_element_by_xpath(book_a_tee_xpath).click()

# Get dates row
date_row = driver.find_element_by_xpath("//tr")
dates_str_list = [i.text.split('\n')[1] for i in date_row.find_elements_by_xpath('./child::*')[1:]]

now = datetime.now()

dates_list = list()
for date_string in dates_str_list:
    date = date_string.split()[0]
    month = date_string.split()[1]

    year = str(now.year)
    if (now.month == 12) and (month == 'Jan'):
        year = str(now.year + 1)

    dates_list.append(datetime.strptime(year + month + date, "%Y%b%d").date())

# Find index of correct date
today_8pm = datetime.strptime(str(now).split()[0] + '20', "%Y-%m-%d%H")
days = 9.5
if datetime.now().hour >= 20:
    days = 10.5
time_diff = timedelta(days=days)
booking_date = (today_8pm + time_diff).date()

booking_index = dates_list.index(booking_date)

# Get club row
club_rows = driver.find_elements_by_xpath("//tr[@class='club-row']")

row_to_use = -1
for i in range(len(club_rows)):
    row_cells = club_rows[i].find_elements_by_xpath('./child::*')
    if row_cells[0].text == 'Waitemata':
        row_to_use = i

waitemata_row = club_rows[row_to_use]

# Click on correct date
booking_elements = [el for el in waitemata_row.find_elements_by_xpath('./child::*') if len(el.text.split()) != 1]

# # # # # # Testing # # # # # # # #
# booking_index = 11

booking_elements[booking_index].click()

# ----------------- Loop until bookings open -----------------#
first_time_xpath = "//table[@id='MainContent_TimeslotsTable']//thead//tbody//tr"
booking_element_xpath = "//td[@class='xavail']"


def is_booking_open(driver):
    driver.refresh()
    time_xpath = "//table[@id='MainContent_TimeslotsTable']//thead//tr//th"
    time_element = driver.find_element_by_xpath(time_xpath)
    return time_element.text == 'Time'


sleep_time = 600
while not is_booking_open(driver):
    if datetime.now().hour == 19:
        sleep_time = 300
        if datetime.now().minute > 50:
            sleep_time = 60
            if datetime.now().minute >= 59:
                sleep_time = 10
                if datetime.now().second > 50:
                    sleep_time = 1
    print("Checking every " + str(sleep_time) + " seconds")
    time.sleep(sleep_time)

first_time_xpath = "//table[@id='MainContent_TimeslotsTable']//tbody//tr"
first_time_row = driver.find_element_by_xpath(first_time_xpath)

if all(el.get_attribute('class') == 'xavail' for el in first_time_row.find_elements_by_xpath("./child::*")[2:]):
    first_time_row.find_element_by_xpath(booking_element_xpath).click()
else:
    print('Too Late')


#________________________Test________________________________
first_time_row = driver.find_elements_by_xpath(first_time_xpath)[3]

if all(el.get_attribute('class') == 'xavail' for el in first_time_row.find_elements_by_xpath("./child::*")[2:]):
    first_time_row.find_elements_by_xpath("./child::*")[2].click()
else:
    print('Too Late')
#______________________--------________-----_____------___------______--


# Add other players
def get_full_name(driver, name):
    for i in driver.find_elements_by_xpath("//select[@class='tb-dll-friend form-control']/option"):
        if name in i.text:
            return i.text


i = 1
for player in other_players:
    friend_option_xpath = "//select[@class='form-control ddl-player-type']/option[text()='Friend']"
    driver.find_elements_by_xpath(friend_option_xpath)[i].click()

    full_name = get_full_name(driver, player)

    select_friend_xpath = "//select[@class='tb-dll-friend form-control']/option[text()='" + full_name + "']"
    driver.find_elements_by_xpath(select_friend_xpath)[i].click()
    i += 1

# Select number of holes
if number_of_holes == 9:
    nine_check_box_xpath = "//input[@value='NineRadio']"
    nine_check_boxes = driver.find_elements_by_xpath(nine_check_box_xpath)
    for check_box in nine_check_boxes:
        try:
            check_box.click()
        except:
            pass

# Click submit
submit_xpath = "//input[@value='SUBMIT']"
driver.find_element_by_xpath(submit_xpath).click()
```
